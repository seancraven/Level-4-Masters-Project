```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['font.size'] = 18
```

```python
#Load Data 
min_mape_10_3 = np.loadtxt('./Training_Data/Guassian_noise0.0001-0.1mape_103.csv',delimiter = ',')
bool_10_3     = np.loadtxt('./Training_Data/Guassian_noise0.0001-0.1bool_103.csv',delimiter = ',')


bool_10_5     = np.loadtxt('./Training_Data/Guassian_noise0.0001-0.1bool.csv',delimiter = ',')
min_mape_10_5 = np.loadtxt('./Training_Data/Guassian_noise0.0001-0.1mape.csv',delimiter = ',')

numerical_mape = np.loadtxt('./Training_Data/analytic_mape.csv',delimiter=',')

noise_sigma = np.logspace(-4,-1,10)
```

```python
mean_min_mape_10_5 = np.mean(min_mape_10_5,axis= 1)
ste_10_5 = np.std(min_mape_10_5,axis= 1)/min_mape_10_5.shape[1]**0.5
mean_min_mape_10_3 = np.mean(min_mape_10_3,axis= 1)
ste_10_3 = np.std(min_mape_10_3,axis= 1)/min_mape_10_3.shape[1]**0.5
```

```python
fig,ax = plt.subplots(1,figsize = (10,6))

ax.errorbar(noise_sigma, mean_min_mape_10_3,ste_10_3,c = 'purple', capsize = 2, label = (r'$V(\phi)_{{{nn}}}:10^3$ Training Datapoints '))
ax.errorbar(noise_sigma, mean_min_mape_10_5,ste_10_5,c = 'blue' , capsize = 2, label = (r'$V(\phi)_{{{nn}}}:10^5$ Training Datapoints '))
ax.plot(noise_sigma,numerical_mape,linestyle = '--',c = 'black',label = '$V(\phi):$ Numeric')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylabel('Mean Minumum MAPE')
ax.set_xlabel('$\sigma$')
ax.legend()
fig.savefig('../Figures/mape(sigma)_with_varying_noise.png',dpi= 300)
```
