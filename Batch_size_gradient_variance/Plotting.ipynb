```python
import numpy as np
import matplotlib.pyplot as plt
import os 
plt.rcParams.update({
    "text.usetex": True})
```

```python
path = '/home/sean/Documents/Work/Level 4/Level-4-Masters-Project/Batch_size_gradient_variance/Weights_Data'
file_names = os.listdir(path)
batch_sizes = [4,16,32,64,128,256]
no_layers = 6
fig, ax = plt.subplots(len(batch_sizes),no_layers, figsize = (10,10), sharex = True)
epochs  = 100
for i,batch_size in enumerate(batch_sizes):
    stds  = np.loadtxt('/home/sean/Documents/Work/Level 4/Level-4-Masters-Project/Batch_size_gradient_variance/Weights_Data/batch_size_{}std.csv'.format(batch_size),delimiter= ',') 
    means = np.loadtxt('/home/sean/Documents/Work/Level 4/Level-4-Masters-Project/Batch_size_gradient_variance/Weights_Data/batch_size_{}rms.csv'.format(batch_size),delimiter= ',')
    for j in range(no_layers):
        sigma_plus = means[:,j]+stds[:,j]
        sigma_minus = means[:,j]-stds[:,j]
        ax[i,j].fill_between(range(epochs),sigma_plus,sigma_minus,alpha =0.2)
        ax[i,j].plot(range(epochs),means[:,j])
        ax[0,j].set_title(r'Layer {}'.format(j+1))
    ax[i,0].set_ylabel(r'Batch Size {}'.format(batch_size)) 
```
