# Plotting $\vec{W}$ and $\Delta \vec{W}$

Takes the weight data from callback on end of epoch on network trained in weight_batches_4-256.py

```python
import numpy as np
import matplotlib.pyplot as plt
import os 
plt.rcParams.update({
    "text.usetex": True})
```

```python
batch_sizes = [2**j for j in range(4,10)]
epochs = 10
no_layers = 5
```

```python
path = '/home/sean/Documents/Work/Level 4/Level-4-Masters-Project/Batch_size_gradient_variance/Weights_Data/weight'
file_names = os.listdir(path)


fig, ax = plt.subplots(len(batch_sizes),no_layers, figsize = (10,10), sharex = True,sharey = False)

for i,batch_size in enumerate(batch_sizes):
    stds  = np.loadtxt(os.path.join(path,'weigth_batch_size_{}std.csv'.format(batch_size)),delimiter= ',') 
    means = np.loadtxt(os.path.join(path,'weigth_batch_size_{}rms.csv'.format(batch_size)),delimiter= ',')
    for j in range(no_layers):
        sigma_plus = means[:,j]+stds[:,j]
        sigma_minus = means[:,j]-stds[:,j]
        ax[i,j].fill_between(range(epochs),sigma_plus,sigma_minus,alpha =0.2)
        ax[i,j].plot(range(epochs),means[:,j])
        ax[0,j].set_title(r'Layer {}'.format(j+1))
    ax[i,0].set_ylabel(r'Batch Size {}'.format(batch_size)) 
fig.suptitle(r'$E(|\vec{W}|)$',fontsize = 20)
```

```python
path = '/home/sean/Documents/Work/Level 4/Level-4-Masters-Project/Batch_size_gradient_variance/Weights_Data/delta'
#file_names = os.listdir(path)
fig, ax = plt.subplots(len(batch_sizes),no_layers, figsize = (20,20), sharex = True,sharey = True)

for i,batch_size in enumerate(batch_sizes):
    stds  = np.loadtxt(os.path.join(path,'weight_delta_batch_size_{}std.csv'.format(batch_size)),delimiter= ',') 
    means = np.loadtxt(os.path.join(path,'weight_delta_batch_size_{}rms.csv'.format(batch_size)),delimiter= ',')
    for j in range(no_layers):
        sigma_plus = means[:,j]+stds[:,j]
        sigma_minus = means[:,j]-stds[:,j]
        ax[i,j].fill_between(range(epochs),sigma_plus,sigma_minus,alpha =0.2)
        ax[i,j].plot(range(epochs),means[:,j])
        ax[0,j].set_title(r'Layer {}'.format(j+1))
        ax[-1,j].set_xlabel('Training Epochs')
    ax[i,0].set_ylabel(r'Batch Size {}'.format(batch_size)) 
fig.suptitle(r'$E(|\Delta \vec{W}|)$',fontsize = 20)
```
