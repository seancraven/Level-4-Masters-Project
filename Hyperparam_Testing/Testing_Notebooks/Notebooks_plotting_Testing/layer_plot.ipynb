# Layer Depth and Node Count Plotting Notebook
Notebook which takes data from the Layers_testing.ipynb notebook and plots in better format

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'legend.fontsize': 16})
matplotlib.rcParams.update({'font.size': 16})
matplotlib.rcParams.update({'text.usetex':True})
```

```python
min_mape_2d = np.loadtxt('/home/sean/Documents/Work/Level 4/Level-4-Masters-Project/Hyperparam_Testing/Testing_Notebooks/Testing_Data/layer/Min_Mape_of_varying_layer_shapes.csv', delimiter = ',')
min_mape = min_mape_2d.reshape(min_mape_2d.shape[0], min_mape_2d.shape[1]//8,8)
```

```python
layers = np.arange(2,9)
layer_shapes = np.arange(10,610,100)
```

```python
mean_min_mape = np.mean(min_mape,axis = 2)
ste_min_mape = np.std(min_mape,axis=2 )*8**-0.5
mean_min_mape.shape
```

```python
colormap = plt.cm.get_cmap('viridis_r')
colors = [colormap(i) for i in np.linspace(0,1,layer_shapes.shape[0])]
```

```python
fig, axs = plt.subplots(1,1,figsize = (10,6))
for i in range(1,6):
    axs.errorbar(layers,mean_min_mape[i,:],ste_min_mape[i,:], label = str(layer_shapes[i])+' Nodes per Layer' , capsize=2, c = colors[i])
axs.set_ylabel('Mean Minimum MAPE')
axs.set_xlabel('Hidden Layers')
#axs.errorbar(layers,mean_min_mape[5,:],ste_min_mape[5,:])
plt.legend()
plt.savefig('/home/sean/Documents/Work/Level 4/Level-4-Masters-Project/Figures/linechart_layers_20_2_22.png', dpi = 300)
```
